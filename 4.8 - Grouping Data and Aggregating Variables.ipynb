## Question 4 - Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Create the “max_order” column. First, a new column called “max_order” is created,
# which will be what stores the maximum order number for each user (step 3). Then, the ords_prods_merge
# dataframe is grouped by the “user_id” column (step 1). And finally, the transform() function is applied
# on the “order_number” column with the 'max' argument (step 2).

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform('max')
```

```python
# View the head of the dataframe to see the new 'max_order' column.

ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# View counts for each of the three values in the 'loyalty_flag' column.

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

## Question 6 - The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

```python
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spend_type'] = 'Low spender'
```

### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spend_type'] = 'High spender'
```

View the last 50 items in the datafarme to see 'high spender' and 'low spender' values in the 'spend_type' column

```python
ords_prods_merge[['user_id', 'avg_price', 'spend_type']].tail(50)
```

## Question 7 - In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

```python
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

### If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] > 20, 'order_type'] = 'Non frequent customer'
```

### If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] > 10) & (ords_prods_merge['order_frequency'] <= 20), 'order_type'] = 'Regular customer'
```

### If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'order_type'] = 'Frequent customer'
```

### View the first 50 rows to view the different values in the 'order_type' column

```python
ords_prods_merge[['user_id', 'avg_price', 'spend_type', 'order_frequency', 'order_type']].head(50)
```
