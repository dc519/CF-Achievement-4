## Question 2 - Suppose your clients have changed their minds about the labels you created in your busiest_day column. Now, they want Busiest day to become Busiest days (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, theyd also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# First, display counts for each of the days of the week from the 'order_dow' column; Saturday corresponds to 0.

ords_prods_merge['order_dow'].value_counts(dropna = False)
```

```python
# From https://stackoverflow.com/questions/11346283/renaming-column-names-in-pandas -->
# "Renaming column names in Pandas"

ords_prods_merge = ords_prods_merge.rename(columns={'busiest_day': 'busiest_days'})
```

```python
# Viewing the head shows that the name of the column was changed from 'busiest_day' to 'busiest_days'

ords_prods_merge.head()
```

```python
# Here I am inserting the text "Busiest day" if the 'order_dow' is either 0 or 1. The task stated that I needed
# to create a new column for the two slowest days so I am labeling the slow days (days 3 and 4) in the 'busiest_days'
# column with a '-' and then I'm labeling all other days with 'Regularly busy'. I will create the
# 'slowest_days' column in a separate line below.

result = []

for value in ords_prods_merge['order_dow']:
  if value == 0 or value == 1:
    result.append("Busiest day")
  elif value == 3 or value == 4:
    result.append("-")
  else:
    result.append("Regularly busy")
```

```python
# Insert the results from result array into the new column 'slowest_days'

ords_prods_merge['busiest_days'] = result
```

```python
# Here I am saving either the text 'Slowest_day' if the order_dow is either 3 or 4. For all other days I am
# inserting a '-'. I think a better way to do this would be to simply have one column name something like
# 'busy_rate' or something similar and then insert the text 'Busiest day', 'Regularly busy day' or 'Slowest day'
# but the task is likely asking for this approach so that I can understand how to create a new column.

slowest_days_result = []

for value in ords_prods_merge['order_dow']:
  if value == 3 or value == 4:
    slowest_days_result.append('Slowest day')
  else:
    slowest_days_result.append('-')
```

```python
# Insert the results from slowest_days_result array into the new column 'slowest_days'

ords_prods_merge['slowest_days'] = slowest_days_result
```

## Question 4 - When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled Most orders, Average orders, and Fewest orders. Create a new column containing these labels called busiest_period_of_day.

```python
# First, display counts for each of the hours of the day from the 'order_hour_of_day' column.

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# I decided to label the top 5 busiest hours with the text 'Most orders', the bottom 5 least busiest hours
# with the text 'Fewest orders' and all other hours with the text 'Average orders'.

orders_result = []

for value in ords_prods_merge['order_hour_of_day']:
  if value == 10 or value == 11 or value == 14 or value == 15 or value == 13:
    orders_result.append('Most orders')
  elif value == 1 or value == 5 or value == 2 or value == 4 or value == 3:
    orders_result.append('Fewest orders')
  else:
    orders_result.append('Average orders')
```

```python
# Insert the results from orders_result array into the new column 'busiest_period_of_day'.

ords_prods_merge['busiest_period_of_day'] = orders_result
```

```python
# Display the head of the ords_prods_merge dataframe; note that the new column 'busiest_period_of_day'
# has been added and contains text based on the if statements I wrote.

ords_prods_merge.head(15)
```
